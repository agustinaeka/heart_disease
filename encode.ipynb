```python
def encode_sex(sex):
    sex = sex
    if sex == 0:
        sex = 'perempuan'
    else:
        sex = 'laki-laki'
        
    return sex

def encode_cpt(cpt):
    cpt = cpt
    if cpt == 1:
        cpt = 'typical angina'
    elif cpt == 2:
        cpt = 'atypical angina'
    elif cpt == 3:
        cpt = 'non-anginal pain'
    else:
        cpt = 'asymptomatic'
        
    return cpt

def encode_fbs(fbs):
    fbs = fbs
    if fbs == 0:
        fbs = '<120'
    else:
        fbs = '>120'
        
    return fbs

def encode_re(re):
    re = re
    if re == 0:
        re = 'Normal'
    elif re == 1:
        re = 'ST-T Abnormal'
    else:
        re = 'Hipertrophy'
        
    return re

def encode_ea(ea):
    ea = ea
    if ea == 0:
        ea = 'tidak'
    else:
        ea = 'ya'
        
    return ea

def encode_ss(ss):
    ss = ss
    if ss == 1:
        ss = 'upsloping'
    elif ss == 2:
        ss = 'flat'
    else:
        ss = 'downsloping'
        
    return ss

def encode_tr(tr):
    tr = tr
    if tr == 0:
        tr = 'tidak terkena penyakit'
    else:
        tr = 'terkena penyakit'
        
    return tr
```

```python
gender = {'1.Laki-laki': 1, '0.Perempuan': 0}
chest_pain_type = {'1. typical angina': 1,
                   '2. atypical angina': 2,
                   '3. non-anginal pain': 3,
                   '4. asymptomatic ': 4}
fasting_blood_sugar = {'1. > 120': 1, '0. <120': 0}
resting_ecg = {'0. Normal': 0, '1. ST-T Abnormal': 1, '2. Hipertrophy ': 2}
exercise_angina = {'1. Ya': 1, '0.Tidak': 0}
ST_slope = {'1.upsloping': 1, '2.flat': 2, '3.downsloping': 3}
```

```python
data['sex'] = data['sex'].apply(lambda x : encode_sex(x) )
data['chest pain type'] = data['chest pain type'].apply(lambda x : encode_cpt(x) )
data['fasting blood sugar'] = data['fasting blood sugar'].apply(lambda x : encode_fbs(x) )
data['resting ecg'] = data['resting ecg'].apply(lambda x : encode_re(x) )
data['exercise angina'] = data['exercise angina'].apply(lambda x : encode_ea(x) )
data['ST slope'] = data['ST slope'].apply(lambda x : encode_ss(x) )
data['target'] = data['target'].apply(lambda x : encode_tr(x) )
```
